```python
import pandas as pd
df1 = pd.read_csv('/Users/lcdlab/BNC_nltk/Lexical-Complexity-of-Speech-master/adult_utterance.csv')
df2 = pd.read_csv('/Users/lcdlab/BNC_nltk/Lexical-Complexity-of-Speech-master/cleaned_adultToChild.csv')
```

```python
df1 = df1.drop(['Unnamed: 0'], axis = 1)
df2 = df2.drop(['Unnamed: 0'], axis = 1);
```

```python
import nltk
from nltk.tokenize import wordpunct_tokenize
for i in range(df2.shape[0]):
    df2['gloss'].iloc[i] = wordpunct_tokenize(df2['gloss'].iloc[i])
```

```python
for i in range(df2.shape[0]):
    df1['gloss'].iloc[i] = wordpunct_tokenize(df1['gloss'].iloc[i])
```

```python
df1
```

```python
df2
```

```python
from nltk.stem.snowball import SnowballStemmer
ss = SnowballStemmer('english')
df1_new = pd.DataFrame()
for i in range(df1.shape[0]):    
    newrow = df1.iloc[[i]]
    for word in df1['gloss'].iloc[i]:
        if (word!='xxx') & (len(word)>1):
            newrow['gloss'] = ss.stem(word)
            df1_new = df1_new.append(newrow)
```

```python
df2_new = pd.DataFrame()
for i in range(df2.shape[0]):    
    newrow = df2.iloc[[i]]
    for word in df2['gloss'].iloc[i]:
        if (word!='xxx') & (len(word)>1):
            newrow['gloss'] = ss.stem(word)
            df2_new = df2_new.append(newrow)
```

```python
df1_new = df1_cr.drop(['id', 'speaker_id', 'utterance_order', 'transcript_id', 'corpus_id',
       'gloss', 'num_tokens', 'stem', 'part_of_speech', 'speaker_code',
       'speaker_name', 'speaker_role', 'target_child_id',
       'target_child_name', 'target_child_sex', 'type', 'media_end',
       'media_start', 'media_unit', 'collection_id', 'collection_name',
       'num_morphemes', 'language', 'corpus_name'], axis = 1)
df2_new = df2_cr.drop(['id', 'speaker_id', 'utterance_order', 'transcript_id', 'corpus_id',
       'gloss', 'num_tokens', 'stem', 'part_of_speech', 'speaker_code',
       'speaker_name', 'speaker_role', 'target_child_id',
       'target_child_name', 'target_child_sex', 'type', 'media_end',
       'media_start', 'media_unit', 'collection_id', 'collection_name',
       'num_morphemes', 'language', 'corpus_name'], axis = 1)

df1_aoa_ratingmean = df1_new.drop(['Conc.M','SUBTLEX'], axis = 1)
df2_aoa_ratingmean = df2_new.drop(['Conc.M','SUBTLEX'], axis = 1)
df1_cr_concm = df1_new.drop(['Rating.Mean','SUBTLEX'], axis = 1)
df2_cr_concm = df2_new.drop(['Rating.Mean','SUBTLEX'], axis = 1)
df1_cr_subtlex = df1_new.drop(['Rating.Mean','Conc.M'], axis = 1)
df2_cr_subtlex = df2_new.drop(['Rating.Mean','Conc.M'], axis = 1)

df1_aoa_ratingmean = df1_aoa_ratingmean.dropna()
df2_aoa_ratingmean = df2_aoa_ratingmean.dropna()
df1_cr_concm = df1_cr_concm.dropna()
df2_cr_concm = df2_cr_concm.dropna()
df1_cr_subtlex = df1_cr_subtlex.dropna()
df2_cr_subtlex = df2_cr_subtlex.dropna()
```

```python
import numpy as np
import seaborn as sns
par = np.polyfit(df1_aoa_ratingmean['target_child_age'], df1_aoa_ratingmean['Rating.Mean'], 1, full=True)
slope=str(par[0][0])
intercept=str(par[0][1])

sns.regplot(df1_aoa_ratingmean['target_child_age'], df1_aoa_ratingmean['Rating.Mean'])
plt.show()
print('Adult-direct')
print('Slope: ' + slope)
print('Intercept: '+ intercept)
```
