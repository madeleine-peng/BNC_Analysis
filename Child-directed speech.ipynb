```python
import csv
import numpy as np
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
utt = pd.read_csv('/Users/anna/Documents/GitHub/Lexical-Complexity-of-Speech/adultToChild.csv')
```

```python
utt.rename(columns = {'Unnamed: 0':'index'}, inplace = True)
utt['index'] = utt['index']-1
utt.sort_values(['corpus_id', 'transcript_id', 'utterance_order'])
```

```python
utt.shape
```

```python
listutt = list(utt['speaker_role'])
string = 'target_child'
if string in listutt:
    print('True')
```

```python
df = pd.DataFrame()
for i in range(utt.shape[0]-1):
    j = i + 1
    curr_corpus = utt['corpus_id'].iloc[i]
    next_corpus = utt['corpus_id'].iloc[j]
    if curr_corpus == next_corpus:
        curr_trans = utt['transcript_id'].iloc[i]
        next_trans = utt['transcript_id'].iloc[j]
        if curr_trans == next_trans:
            curr = utt['speaker_role'].iloc[i]
            nextr = utt['speaker_role'].iloc[j]
            if ((curr == 'Child' or curr == 'Target_Child') & (nextr != 'Child') & (nextr != 'Target_Child')) or ((nextr == 'Child' or nextr == 'Target_Child') & (curr != 'Child') & (curr != 'Target_Child')):
                curr_order = utt['utterance_order'].iloc[i]
                next_order = utt['utterance_order'].iloc[j]
                if next_order - curr_order == 1:
                    df = df.append(utt.iloc[[i]])
                    df = df.append(utt.iloc[[j]])
```

```python
df = df.drop_duplicates()
df.shape
```

```python
index = df['index']
index1 = index[::2]
index2 = index[1::2]
index_df = range(df.shape[0])
odd = index_df[::2]
even = index_df[1::2]
```

```python
n = 0
for k in index1:
    i = odd[n]
    n += 1
    j = k - 1
    if j >= 0:
        curr_corpus = df['corpus_id'].iloc[i]
        next_corpus = utt['corpus_id'].iloc[j]
        if curr_corpus == next_corpus:
            curr_trans = df['transcript_id'].iloc[i]
            next_trans = utt['transcript_id'].iloc[j]
            if curr_trans == next_trans:
                curr_speaker = df['speaker_id'].iloc[i]
                next_speaker = utt['speaker_id'].iloc[j]
                if curr_speaker == next_speaker:
                    curr_order = df['utterance_order'].iloc[i]
                    next_order = utt['utterance_order'].iloc[j]
                    if next_order - curr_order == -1:
                        df['num_tokens'].iloc[i] += utt['num_tokens'].iloc[j]
n = 0
for k in index2:
    i = even[n]
    n += 1
    j = k - 1
    if j < utt.shape[0]:
        curr_corpus = df['corpus_id'].iloc[i]
        next_corpus = utt['corpus_id'].iloc[j]
        if curr_corpus == next_corpus:
            curr_trans = df['transcript_id'].iloc[i]
            next_trans = utt['transcript_id'].iloc[j]
            if curr_trans == next_trans:
                curr_speaker = df['speaker_id'].iloc[i]
                next_speaker = utt['speaker_id'].iloc[j]
                if curr_speaker == next_speaker:
                    curr_order = df['utterance_order'].iloc[i]
                    next_order = utt['utterance_order'].iloc[j]
                    if next_order - curr_order == 1:
                        df['num_tokens'].iloc[i] += utt['num_tokens'].iloc[j] 
```

```python
df = df[df.gloss != 'xxx']
df.shape
```

```python
for i in range(len(df['index'])):
    string = df['gloss'].iloc[i]
    if 'xxx' in str(string):
        count = str(string).count('xxx')
        df['num_tokens'].iloc[i] -= count
```

```python
df = df[pd.notnull(df['gloss'])]
df.shape
```

```python
df = df[(df.speaker_role != 'Child') & (df.speaker_role != 'Target_Child')]
```
